## **Example: Polarity Analysis**
The most accessible way to start with sentiment analysis in Python is using the TextBlob library. It uses a pre-trained model to provide a polarity score (ranging from -1.0 to 1.0).

How to read the results:
- Polarity > 0: The text expresses a positive sentiment.
- Polarity < 0: The text expresses a negative sentiment.
- Polarity = 0: The text is objective or neutral.

```python
def analyze_feedback(text):
    # Creating a TextBlob object:
    analysis = TextBlob(text)

    # Polarity is a float between -1.0 (very negative) and 1.0 (very positive):
    polarity = analysis.sentiment.polarity

    # Classifying the result:
    if polarity > 0:
        category = 'Positive'
    elif polarity < 0:
        category = 'Negative'
    else:
        category = 'Neutral'

    return category, polarity

# Test Data:
reviews = [
    'I absolutely love this new smartphone! The camera is incredible.',
    'The service was okay, but the food was a bit cold.',
    'This is the worst experience I have ever had with a brand.'
]

print(f'{'Review':<60} | {'Sentiment':<10} | {'Score'}')
print('-' * 85)

for r in reviews:
    sentiment, score = analyze_feedback(r)
    print(f'{r:<60} | {sentiment:<10} | {score:.2f}')
```
