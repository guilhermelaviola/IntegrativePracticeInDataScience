# **Treatment, Validation & Testing**
In Machine Learning, model validation and testing are essential for ensuring the reliability and accuracy of predictions. Various metrics such as accuracy, precision, recall, and F1-Score evaluate model performance, with each metric serving specific objectives. Accuracy reflects the overall correctness of classifications but can be misleading in class-imbalanced scenarios. Precision measures the accuracy of positive predictions, while recall assesses the identification of true positives, focusing on minimizing false negatives. The F1-Score, a harmonic mean of precision and recall, offers a balance between the two. Additionally, the confusion matrix provides a visual representation of model performance, highlighting true positives, true negatives, false positives, and false negatives. Data quality significantly influences model performance, and addressing anomalies and noise through preprocessing techniques is critical for accurate predictions. Thus, careful selection of metrics, confusion matrix analysis, and effective data preprocessing are vital for developing robust Machine Learning models.

```python
# Importing all the necessary libraries and resources:
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

## **Example: Loading Data**
Below we load an Iris dataset, we converting it to a binary classification (Setosa vs non-Setosa), and we simulate class imbalance by removing many positive samples. Finally, we split the data into training and test sets.

```python
# Iris dataset:
iris = load_iris()
X, y = iris.data, iris.target

# Converting it to a binary classification:
y = (y == 0).astype(int)

# Simulating class imbalance:
idx = np.where(y == 1)[0]
drop = np.random.choice(idx, size=int(len(idx) * 0.7), replace=False)
X = np.delete(X, drop, axis=0)
y = np.delete(y, drop, axis=0)

# Splitting the data:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## **Example: Preprocessing Data**
Below, we first standardize the training and test features so they have mean 0 and standard deviation 1, which helps the model perform better. Then we create and train a logistic regression classifier using the scaled training data. Finally, we use the trained model to predict class labels for the scaled test data.

```python
# Scaling:
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Logistic Regression model:
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions:
y_pred = model.predict(X_test)
```

## **Example: Metrics**
Below we define variables to store the metrics such as Accuracy, Precision, Recall and F1-Score, then we display the results.

```python
# Defining the metrics:
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
cm = confusion_matrix(y_test, y_pred)
```

```python
print('=== Model Evaluation Metrics ===')
print(f'Accuracy: {accuracy:.3f}')
print(f'Precision: {precision:.3f}')
print(f'Recall: {recall:.3f}')
print(f'F1-Score: {f1:.3f}')

print('\n=== Confusion Matrix ===')
print(cm)
```
