# **Decision-making**
Decision-making is an iterative, structured process of selecting the best course of action from various alternatives—specifically action, inaction, or deferral—to achieve a defined objective. It evolves through six key stages: defining the problem, gathering information, evaluating alternatives based on specific criteria (such as cost, risk, and ethics), making and implementing the choice, and finally assessing the results. For complex scenarios, Multicriteria Decision Making provides a quantitative framework to rank options by weighing multiple competing factors. Ultimately, a successful decision depends on balancing available data with situational needs to ensure the chosen path offers the best possible outcome for all parties involved.

## **Example: Multicriteria Decision Making (MCDM)**
To exemplify this, we can use a Weighted Scoring Model. This Python script evaluates three potential software vendors based on four criteria: Cost, Features, Support, and Risk.

```python
# Defining the Criteria and their relative Importance (Weights must sum to 1.0):
criteria_weights = {
    'cost': 0.40,        # Lower is usually better, but here we score 1-10 (10 = most affordable)
    'features': 0.30,    # 10 = has everything we need
    'support': 0.20,     # 10 = 24/7 dedicated support
    'risk': 0.10         # 10 = very stable/low risk
}

# Defining the Alternatives and their scores (1 to 10 scale) for each criterion:
alternatives = {
    'Vendor_A': {'cost': 8, 'features': 6, 'support': 9, 'risk': 7},
    'Vendor_B': {'cost': 5, 'features': 9, 'support': 7, 'risk': 8},
    'Vendor_C': {'cost': 6, 'features': 7, 'support': 6, 'risk': 9}
}
```

```python
# Calculating the weighted total for each alternative:
def evaluate_decisions(options, weights):
    results = {}

    for name, scores in options.items():
        total_score = sum(scores[c] * weights[c] for c in weights)
        results[name] = round(total_score, 2)

    return results
```

```python
# Making a Decision based on the highest score:
final_scores = evaluate_decisions(alternatives, criteria_weights)
best_choice = max(final_scores, key=final_scores.get)
```

```python
# Output results:
print('--- Decision Evaluation Results ---')
for vendor, score in sorted(final_scores.items(), key=lambda x: x[1], reverse=True):
    print(f'{vendor}: {score}')

print(f'\nConclusion: The best decision is to proceed with {best_choice}.')
```
